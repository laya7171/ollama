```python
invetory_db = {
    "laptop": {"stock": 10, "price": 1000},
    "phone": {"stock": 20, "price": 500},
    "keyboard": {"stock": 15, "price": 100}
}
```

```python
# tool 1
def check_inventory(product_name):
    prodct_name  = product_name.lower()

    if product_name in invetory_db:
        return invetory_db[product_name]

    return {"stock": 0, "price": 0}

#tool 2


def check_loyalty(base_price, year_as_customer):
    discount = min(year_as_customer * 0.05, 0.25)  # 5% discount per year, up to 25%
    final_price = base_price * (1 - discount)
    return round(final_price, 2)
```

```python
availabe_functions ={
    'check_inventory': check_inventory,
    'check_loyalty': check_loyalty
}
```

```python
tools = [
    {
        "type": "function",
        "function":{
            "name": "check_inventory",
            "description": "Check the inventory and price of a product.",
            "parameters": {
                "type": "object",
                "properties": {
                    "type": "string",

            },
                "required": ["product_name"]
        }
    },
        "type": "function",
        "function":{
            "name": "check_loyalty",
            "description": "Calculate the final price after applying loyalty discount.",
            "parameters": {
                "type": "object",
                "properties": {
                    "base_price": {"type": "number"},
                    "year_as_customer": {"type": "number"}
                },
                "required": ["base_price", "year_as_customer"]
            }
        }

    }
]
```

```python
tools
```

```python
message = [ {"role": "user",  "content": "I want to buy a laptop, how many are in the stock"}]
```

```python
response1 = ollama.chat(model = "phi4-mini:latest", messages = message, tools = tools, stream= True)
```

```python
for chunk in response1:
    print(chunk['message'], end='', flush=True)
```
