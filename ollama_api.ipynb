```python
import json
import requests

url = "http://localhost:11434/api/generate"

payload = {
    "model" : "gemma3:4b",
    "prompt" : "What is a black hole?"
}

response = requests.post(url, json=payload)
```

```python
output = ""
for i in response.iter_lines():
    if i:
        data = json.loads(i.decode('utf-8'))
        output += data.get('response', '')
```

```python
print(output)
```
