```python
import ollama
from langchain_ollama import OllamaLLM
from langchain_core.prompts import PromptTemplate
```

```python
llm = OllamaLLM(model="gemma3:4b")
```

```python
response = llm.invoke("What is a black hole?")
```

```python
templete  = PromptTemplate(
    template = "What is the capital of {country}?",
    input_variables = ["country"]
)

templete2 = PromptTemplate(
    template = "Name a thing to do in this city: {city}",
    input_variables = ["city"]
)
```

```python
chain = templete | llm | templete2 | llm
```

```python
response = chain.invoke({"country": "France"})
```

```python
print(response)
```
